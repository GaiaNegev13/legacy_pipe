```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
combined_df = pd.read_pickle('/home/gaia/Projects/legacy_data/combined_gm_volumes.pkl')
combined_df = combined_df[(combined_df['classification_label'] == 1) | (combined_df['source'] == 'snbb')]

scans = combined_df.drop_duplicates(subset=['subject_id', 'session_id'])
```

```python
# tiv outliers

# plot the tiv of scans by source
plt.figure(figsize=(10, 6))
sns.boxplot(x='source', y='tiv', data=scans)
plt.title('Total Intracranial Volume (TIV) by Source')
plt.xlabel('Source')
plt.ylabel('Total Intracranial Volume (TIV)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("Subjects with TIV under 1000:")
tiv_under_1000 = scans[scans['tiv'] < 1000][['subject_id', 'session_id', 'source', 'tiv']]
```
